<h1 style="color: red">Polynomial Regression</h1>

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

sales = pd.read_csv('./sales_dataset/Advertising.csv')
```

```python
X = sales.iloc[:, 0:4].values
Y = sales.iloc[:, 4:].values
```

```python
X
```

```python
plt.plot(X, Y, 'b.', label="Advertising")
plt.xlabel("Media")
plt.ylabel("Sales")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
deg = 3;
ply_ftres = PolynomialFeatures(degree=deg, include_bias=False)
x_poly = ply_ftres.fit_transform(x_train)
```

```python
ply_reg = LinearRegression()
ply_reg.fit(x_poly, y_train)
```

```python
x_tst_ply = ply_ftres.transform(x_test)
y_pred = ply_reg.predict(x_tst_ply)
```

```python
y_pred
```

```python
y_test
```

```python
# MSE loss
def mse(t1, t2):
    diff = t1 - t2
    return np.sum(diff * diff) / len(diff)
```

```python
loss = mse(y_pred,y_test)
loss
```
